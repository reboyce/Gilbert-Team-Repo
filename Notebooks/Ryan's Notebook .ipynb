Creating a DataFrame with Title and Popularity 

```python
title_pop = tmdb1[['title', 'popularity']]
title_pop
```

```python
title_pop2 = title_pop.loc[title_pop['popularity']== 0.600].count()
title_pop2
```

```python
title_pop30 = title_pop.loc[title_pop['popularity']>=39.293].sort_values('popularity', ascending=False)
title_pop30
```

```python
import numpy as np
import matplotlib.pyplot as plt

#creating dataset
data = title_pop30
courses = title_pop30['title']
values = title_pop30['popularity']
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values,
        width = 0.4)
 
plt.xlabel('Movie Title')
plt.ylabel("Popularity")
plt.title("Movie Popularity by Title")
plt.xticks(rotation = 90)
plt.show()
```
